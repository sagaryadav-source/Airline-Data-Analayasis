# exploratory of the data analysis

# AIRLINE

```python
plt.figure(figsize =(9,10))
ax = sns.countplot(data = df,x="airline")
ax.bar_label(ax.containers[0])
plt.title("count of the airline")
plt.xticks(rotation =60)
plt.show()
```

```python
gp =df.groupby("airline").agg({"airline":"count"})
plt.pie(gp["airline"],labels =gp.index,autopct="%1.2f%%")
plt.title("percentage of the Airline count")
plt.show()
```

```python
plt.figure(figsize=(9,8))
colours = ['red','blue','green','yellow','pink','brown']
sa= sns.histplot(data=df,x="destination_city",palette= colours)
sa.bar_label(sa.containers[0])
plt.title("count of the destination city")
plt.xticks(rotation = 60)
plt.show()
```

```python
plt.figure(figsize =(9,10))
ax = sns.countplot(data = df,x="arrival_time")
ax.bar_label(ax.containers[0])
plt.title("count of the arrival_time in flights")
plt.xticks(rotation = 60)
plt.show()
```
